```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Label', 'Target'], axis=1)
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Creating a PipeLine that contains both a StandardScaler and a KNN model
scaler = StandardScaler()
knn = KNeighborsClassifier()
```

```python
operations = [('scaler', scaler), ('knn', knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(operations)
```

- Performing a grid-search with the pipeline to test various values of k and report back the best performing parameters.

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values = list(range(1, 30))
```

```python
parameters = {'knn__n_neighbors': k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe, parameters, cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
# check best estimator
full_cv_classifier.best_estimator_.get_params()
```

- Using the .cv_results_

```python
pd.DataFrame(full_cv_classifier.cv_results_).head()
```

```python
# plot of the mean test scores per K value
mean_test_scores = full_cv_classifier.cv_results_['mean_test_score']
```

```python
mean_test_scores
```

```python
plt.plot(k_values, mean_test_scores, 'o-')
plt.xlabel('K')
plt.ylabel('Mean Test Score / Accuracy');
```

- Using the grid classifier object from the previous step, get a final performance classification report and confusion matrix.

```python
full_pred = full_cv_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test, full_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(full_cv_classifier, X_test, y_test)
```

- As we can see from the confusion matrix, there are 1 False Positive and 1 False Negative.

```python
print(classification_report(y_test, full_pred))
```
