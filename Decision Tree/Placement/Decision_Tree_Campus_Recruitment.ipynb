```python
#dependent and independent features - For Feature Selection using RFECV
X = placement.drop(['status'], axis=1)
y = placement['status']
```

```python
#Recursive Feature Selection
rfe = RFECV(DecisionTreeClassifier(), cv=StratifiedKFold(3), step=1)
```

```python
rfe.fit(X, y)
```

```python
#optimal number of features
rfe.n_features_
```

```python
#feature importance with ranking
rfe.ranking_
```

```python
#the features that have significance for the classifier
rfe.get_support()
```

```python
#selected Features
feature_selected = X.columns[rfe.get_support()]
feature_selected
```

```python
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train = rfe.transform(X_train)
X_test = rfe.transform(X_test)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

cf = classification_report(y_test, pred)
print(cf)
```
