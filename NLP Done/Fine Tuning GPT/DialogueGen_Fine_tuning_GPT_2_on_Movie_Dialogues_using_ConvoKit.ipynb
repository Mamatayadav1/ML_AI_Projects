```python
from datasets import load_dataset
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments, DataCollatorForLanguageModeling
```

```python
# Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

```python
# Load your text dataset
dataset = load_dataset("text", data_files="cornell_gpt2_data.txt")
```

```python
# Tokenize
def tokenize(example):
    return tokenizer(example["text"], truncation=True, max_length=128, padding="max_length")
```

```python
tokenized = dataset.map(tokenize, batched=True)
```

```python
# Training args
args = TrainingArguments(
    output_dir="./gpt2-cornell",
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=2,
    save_steps=500,
    save_total_limit=2,
    logging_steps=100
)
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
# Trainer
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized["train"],
    tokenizer=tokenizer,
    data_collator=data_collator,
)
```

```python
# (1) Use smaller dataset
tokenized = dataset["train"].select(range(5000)).map(tokenize, batched=True)

# (2) Use max_steps
args = TrainingArguments(
    output_dir="./gpt2-cornell",
    overwrite_output_dir=True,
    max_steps=1000,  # Limit training to 1000 steps
    per_device_train_batch_size=4,
    logging_steps=100
)
```

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
prompt = (
    "Person A: Hey, I can't do this anymore.\n"
    "Person B: Why not?\n"
    "Person A:"
)
```

```python
outputs = generator(
    prompt,
    max_new_tokens=100, # More tokens for multi-line output
    num_return_sequences=3, # Generate 3 variations
    temperature=0.9,  # Creative diversity
    top_p=0.95,  # Nucleus sampling
    do_sample=True,# Enable sampling
    truncation=True
)
```

```python
for i, out in enumerate(outputs):
    print(f"\n--- Generated #{i+1} ---\n")
    for line in out['generated_text'].split("\n"):
        if line.strip():
            print(line.strip())
```
